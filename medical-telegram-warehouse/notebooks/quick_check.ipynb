# Quick Data Check

Quick notebook to verify data is loaded and models are working.

```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()
DATABASE_URL = os.getenv("DATABASE_URL", "postgresql://postgres:postgres@localhost:5432/medical_warehouse")
engine = create_engine(DATABASE_URL)

print("✓ Database connection established")
```

```python
# Check all tables
tables = {
    "Raw": "raw.telegram_messages",
    "Staging": "stg_messages",
    "Channels": "dim_channels",
    "Dates": "dim_dates",
    "Fact": "fct_messages"
}

for name, table in tables.items():
    try:
        count = pd.read_sql(f"SELECT COUNT(*) as count FROM {table}", engine)
        print(f"✓ {name}: {count['count'].iloc[0]} rows")
    except Exception as e:
        print(f"✗ {name}: Error - {e}")
```

```python
# Sample data from fact table
print("\nSample from fact table:")
sample = pd.read_sql("""
    SELECT 
        fm.message_id,
        dc.channel_name,
        LEFT(fm.message_text, 50) as message_preview,
        fm.view_count,
        fm.product_type
    FROM fct_messages fm
    JOIN dim_channels dc ON fm.channel_key = dc.channel_key
    LIMIT 5
""", engine)
display(sample)

print("\n✓ All checks complete!")
```
